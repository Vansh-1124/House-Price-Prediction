```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn import ensemble

clf = ensemble.GradientBoostingRegressor(
    n_estimators=400,
    max_depth=5,
    min_samples_split=2,
    learning_rate=0.1,
    loss='squared_error'  # Corrected loss function
)

clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
import numpy as np

# Define params dictionary
params = {
    'n_estimators': 400,  # Example value
}

# Use the parameter correctly
t_sc = np.zeros((params['n_estimators']), dtype=np.float64)
print(t_sc)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error  # Import loss function

t_sc = np.zeros(len(list(clf.staged_predict(x_test))), dtype=np.float64)  # Initialize loss array

for i, y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i] = mean_squared_error(y_test, y_pred)  # Use MSE as loss function

print(t_sc)  # Output loss at each stage
```

```python
testsc = np.arange((params['n_estimators']))+1
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca.fit_transform(scale(train1))
```

```python
# Print R-squared value for Linear Regression
print("Linear Regression Model R^2 score: ", reg.score(x_test, y_test))
```

```python
# Print R-squared value for Gradient Boosting Regressor
print("Gradient Boosting Regressor Model R^2 score: ", clf.score(x_test, y_test))
```

```python
# Plotting training vs testing error for Gradient Boosting Regressor
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc, clf.train_score_, 'b-', label='Training Error')
plt.plot(testsc, t_sc, 'r-', label='Test Error')
plt.title('Training vs Testing Error for Gradient Boosting')
plt.xlabel('Number of Estimators')
plt.ylabel('Error')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# Predict using both models
y_pred_lr = reg.predict(x_test)
y_pred_gb = clf.predict(x_test)

# Calculate MSE for Linear Regression
mse_lr = mean_squared_error(y_test, y_pred_lr)
print("Linear Regression Mean Squared Error: ", mse_lr)

# Calculate MSE for Gradient Boosting
mse_gb = mean_squared_error(y_test, y_pred_gb)
print("Gradient Boosting Mean Squared Error: ", mse_gb)
```

```python
# PCA for dimensionality reduction
pca = PCA(n_components=2)  # Reduce to 2 components for visualization
principal_components = pca.fit_transform(scale(train1))

# Create a DataFrame with the PCA results
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Price'] = labels  # Add price as the target variable for visualization

# Plot the PCA components
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Price', palette='viridis', data=pca_df)
plt.title('PCA of Features')
plt.show()
```
